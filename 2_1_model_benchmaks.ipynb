### Проверка доступности моделей и возможности их загрузки на мой компьютер

```python
from sentence_transformers import SentenceTransformer
import gc
import time

# чистим кэш, чтобы измерить вместе с временем загрузки
!rm -rf ~/.cache/huggingface/hub

load_times = {}

for name, model_id in MODELS.items():
    print("--------------------------------------------------")
    print(f"Загрузка модели: {name} ({model_id})\n")
    t0 = time.perf_counter()
    model = SentenceTransformer(model_id, device="cpu")
    _ = model.encode(["тест"], show_progress_bar=False)
    t1 = time.perf_counter()
    load_time = round(t1 - t0, 2)
    load_times[name] = load_time
    del model
    gc.collect()
```

### Выводы

Всё в порядке, модели загружаются, и время приемлемое.

### Проверка скорости создания эмбеддингов

```python
# Проверка одной модели

import torch
#torch.set_num_threads(1)

@torch.inference_mode()
def embed_with_timing(model_id, texts):
    print(f'Загружаем модель {model_id}')
    t0_load = time.perf_counter()
    m = SentenceTransformer(model_id)  # Если передавать явным образом device="cpu", то на моем компьютере encode виснет! Оставляем значение по умолчанию.
    max_seq_len = m.max_seq_length
    t1_load = time.perf_counter()

    print("Замеряем время эмбеддинга")
    t0 = time.perf_counter()
    _ = m.encode(
        texts,
        batch_size=64,
        convert_to_numpy=True,
        normalize_embeddings=True,
        show_progress_bar=True)
    t1 = time.perf_counter()

    total_time = t1 - t0
    del m
    gc.collect()

    return {
        "model_id": model_id,
        "model_max_seq_length": max_seq_len,
        "load_time_sec": round(t1_load - t0_load, 2),
        "total_time_sec": round(total_time, 2),
    }
```

Испытаем на одной модели и на маленьком датасете

```python
requests_sample=df['request'].sample(n=1000, random_state=0).astype(str).tolist()
```

```python
import os
# os.environ["TOKENIZERS_PARALLELISM"] = "false"
# os.environ["OMP_NUM_THREADS"] = "1"
# os.environ["MKL_NUM_THREADS"] = "1"
# os.environ["OPENBLAS_NUM_THREADS"] = "1"

summary = embed_with_timing(
    model_id=MODELS["e5-small"],
    texts=requests_sample,

)
```

```python
display(summary)
```

А теперь проверим все модели на полном датасете

```python
results = []

requests_sample=df['request'].sample(n=10000, random_state=0).astype(str).tolist() # Размер выборки 10000 - того же порядка, что и весь датасет

for key in MODELS:
    metrics = embed_with_timing(
        MODELS[key],
        requests_sample,
    )
    results.append(metrics)
```
